```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv("1. Weather Data.csv")
```

```python
data.head()
```

```python
data.tail()
```

```python
data.shape
```

```python
clear_records=data[data["Weather"]=="Clear"]
print(clear_records)
```

```python
count_wind=data[data["Wind Speed_km/h"]==4]
print(count_wind)
```

```python
null_values=data.isnull().sum()
print(null_values)
```

```python
data.rename(columns={"Weather Condition":"Weather_Condition"} ,inplace=True)
print(data)
```

```python
Visibility_mean=data["Visibility_km"].mean()
print(Visibility_mean)
```

```python
wind_visibility = data[(data["Wind Speed_km/h"] > 24) & (data["Visibility_km"] == 25)]
print(wind_visibility)
```

```python
data['Date/Time'] = pd.to_datetime(data['Date/Time'])
```

```python
data.head()
```

```python
mean_values = data.groupby('Weather_Condition').mean()
print(mean_values)
```

```python
data.dtypes
```

```python
fit_data=data[
    (data['Weather_Condition'] == 'Clear') &
    ((data['Rel Hum_%'] > 50) | (data['Visibility_km'] > 40))
]
print(fit_data)
```

```python
snow_conditions=data[data['Weather_Condition'].str.contains('Snow')]
print(snow_conditions)
```

```python
column_counts = snow_conditions.count()
print(column_counts)
```

```python
csv_file_path = '/content/my_data.csv'
```

```python
data.to_csv(csv_file_path, index=False)
print(f"CSV file has been created: {csv_file_path}")
```
